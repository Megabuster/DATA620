```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

edges = [
    ('Andre', 'Carol'),
    ('Andre', 'Fernando'),
    ('Andre', 'Beverly'),
    ('Andre', 'Diane'),
    ('Carol', 'Fernando'),
    ('Carol', 'Diane'),
    ('Diane', 'Fernando'),
    ('Diane', 'Garth'),
    ('Diane', 'Ed'),
    ('Diane', 'Beverly'),
    ('Ed', 'Garth'),
    ('Ed', 'Beverly'),
    ('Fernando', 'Garth'),
    ('Fernando', 'Heather'),
    ('Fernando', 'Garth'),
    ('Garth', 'Heather'),
    ('Heather', 'Ike'),
    ('Ike', 'Jane')
         ]

G = nx.Graph()
G.add_edges_from(edges)
plt.figure(figsize=(10,10))
nx.draw(G, with_labels=True)
plt.show()
```
