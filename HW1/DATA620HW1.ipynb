This is a lab to test our coding environments and deployment systems. NetworkX was chosen to represent the graph for practice for future labs.

The graph used is the sample Krackhardt kite from the assignment instructions.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create an edge list. Only one side of the connection is needed.
edges = [
    ('Andre', 'Carol'),
    ('Andre', 'Fernando'),
    ('Andre', 'Beverly'),
    ('Andre', 'Diane'),
    ('Beverly', 'Garth'),
    ('Carol', 'Fernando'),
    ('Carol', 'Diane'),
    ('Diane', 'Fernando'),
    ('Diane', 'Garth'),
    ('Diane', 'Ed'),
    ('Diane', 'Beverly'),
    ('Ed', 'Garth'),
    ('Ed', 'Beverly'),
    ('Fernando', 'Garth'),
    ('Fernando', 'Heather'),
    ('Fernando', 'Garth'),
    ('Garth', 'Heather'),
    ('Heather', 'Ike'),
    ('Ike', 'Jane')
         ]

# Build and display the graph.
G = nx.Graph()
G.add_edges_from(edges)
layout = nx.spring_layout(G)
nx.draw(G, layout, with_labels=True)
plt.show()
```

The above Krackhardt kite simulates social media connections. The graph is basic in that it only checks if two people are connected without any care for weights or direction.
